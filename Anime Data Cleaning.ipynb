```python
df = df.drop(columns = ['uid','link','synopsis','img_url'])
df
```

```python
df = df.rename( columns = {'title' : 'Anime Name', 'genre' : 'Category', 'aired': 'Release Date and End Date','episodes': 'Episodes',
                           'members': 'Member' ,'popularity': 'Popularity', 	'ranked': 'Ranked', 'score':'Score'}) 
df 
# Changing name columns title more suitable
```

```python
df['Anime Name'] =df['Anime Name'].str.strip('[!@#$%^&*(),.?":{}|<>]"')
df['Category'] =df['Category'].str.strip('[!@#$%^&*(),.?":{}|<>]"')
df = df.drop(columns = 'Remove_Special_Characters')
df 
# remote some special characters out of data
```

```python
df[['Release Date','End Date']]=df['Release Date and End Date'].str.split('to', expand= True)
df = df.drop(columns = 'Release Date and End Date')
df 
# split column Release Date and End Date to easy to see
```

```python
df=df.fillna(' ')
df
```

```python
for x in df.index:                  # checking any mising anime dont have name
    if df.loc[x,"Anime Name"] == '':
        df.drop(x, inplace= True)
df 
#another way to drop null values
#df = df.dropna(subset="Anime Name"), inplace=True)
```

```python
# after ckecking data i still see some special characters in Anime Name cloumn so decice to clear it by other way
# Define a function to remove special characters
def remove_special_characters(text):
    import re
    pattern = r'[^a-zA-Z0-9\s]'  # Keep letters, numbers, and spaces
    return re.sub(pattern, '', text)

# Apply the function to each cell in the DataFrame
df['Anime Name'] = df['Anime Name'].apply(remove_special_characters)
df
```
