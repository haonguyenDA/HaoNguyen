```python
# In your Python script, import the necessary libraries:
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
# Use the requests library to fetch the HTML content of the webpage you want to scrape 
# and arse the HTML content using BeautifulSoup.This makes it easier to navigate and extract data
```

```python
urf ='https://en.wikipedia.org/wiki/List_of_companies_of_Vietnam'

page = requests.get(urf)

soup = BeautifulSoup(page.text,'html')
```

```python
print (soup)
```

```python
# Locate Data:
#Use the methods provided by BeautifulSoup to locate the specific data you want to scrape. 
#This often involves inspecting the HTML structure of the webpage
```

```python
soup.find_all('table')[1]
```

```python
table=soup.find_all ('table')[1]
```

```python
print (table)
```

```python
table.find_all ('th')
```

```python
rows_titles = table.find_all ('th')
```

```python
print (rows_titles)
```

```python
# Select rows titles and pandas print it out
```

```python
table_titles = [title.text.strip() for title in rows_titles]

print (table_titles)
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(columns = table_titles)

df
```

```python
# Identify data input in columns of the table
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data =row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print (individual_row_data)
    length = len(df)
    df.loc[length]= individual_row_data
```
